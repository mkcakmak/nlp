# RegEx in Python

```python
"""RegEx are fully integrated with most search engines and can allow for more robust searching. Nearly all data scientists, especially 
those who work with texts, use RegEx at some stage in their workflow, from data searching, to cleaning data,
to implementing machine learning models. It is an essential tool for any text-based researcher. """

import re

pattern = r"((\d){1,2} (January|February|March|April|May|June|July|August|September|October|November|December))"

text = "This is a date 2 February. Another date would be 14 August."
matches = re.findall(pattern,text)
print (matches)

# The first ( tells RegEx that Im looking for something within the ending )
# we state (\d){1,2}. This means that we are looking for any digit (0-9) that occurs either once or twice ({1,2}).
# (January|February|March|April|May|June|July|August|September|October|November|December)  this indicates another component of the pattern (because it is parentheses). 
# The | indicates the same concept as or in English, so either January, or February, or March, etc
```

```python
# Different result because of "or"

pattern = r"(((\d){1,2}( (January|February|March|April|May|June|July|August|September|October|November|December)))|(((January|February|March|April|May|June|July|August|September|October|November|December) )(\d){1,2}))"

text = "This is a date February 2. Another date would be 14 August."
matches = re.findall(pattern, text)
print (matches)
```

```python
from collections import Counter
import spacy
nlp = spacy.load("en_core_web_md")

corpus = open("atis_utterances.txt", "r").read().split("\n")

all_ent_labels = []


for sentence in corpus:
   sentence = sentence.strip()     # strip --> Remove spaces at the beginning and at the end of the string
   if sentence:
        doc = nlp(sentence)
        ents = doc.ents
        all_ent_labels += [ent.label_ for ent in ents]

c = Counter(all_ent_labels)

print(c)
```

# Extract Location Names

```python
import spacy 
from spacy.matcher import Matcher   
nlp = spacy.load("en_core_web_md")
matcher = Matcher(nlp.vocab)

pattern = [{"POS": "ADP"}, {"ENT_TYPE":"GPE"}]

matcher.add("prepositionLocation",[pattern])

doc = nlp("Show me flights from Denver to Boston on tuesday") 
matches = matcher(doc) 
print(doc.ents)
for mid, start, end in matches: 
    print(doc[start:end]) 
```

# Using Dependency Tree in Semantic Parsing

```python
import spacy

nlp = spacy.load("en_core_web_md")

def reach_parent(source_token, dest_token):
  while source_token != dest_token:
        if source_token.head == source_token:
            return None
        source_token = source_token.head 
        print(source_token)
  return source_token

doc = nlp("I'm going to a conference in Munich.")

print(doc[-2])

print(doc[3])

print(doc[-1], "\n")

#for token in doc:
#    print(token.text, token.head,"\n")

print("\n",reach_parent(doc[-2], doc[3]), "\n")

print(reach_parent(doc[-1], doc[3]))

print(reach_parent(doc[0], doc[3]))


```

```python
# Conjunctions 

import spacy

nlp = spacy.load("en_core_web_md")
doc = nlp("show all flights and fares from denver to san fransisco")

for token in doc:
    if token.dep_ == "dobj":
        print( token.head,token.text,token.conjuncts)
```

```python
# Different Style

import spacy

nlp = spacy.load("en_core_web_md")

doc = nlp("show all flights and fares from denver to san fransisco")

for token in doc:
    if token.dep_ == "dobj":
        dobj = token.text
        conj = [t.text for t in token.conjuncts]
        verb = token.head
        print(verb, dobj, conj)
```

```python
import spacy
nlp = spacy.load("en_core_web_md")

doc = nlp("I want to make a reservation for a flight")


# Extract the direct object and its transitive verb

for token in doc:
    if token.dep_ == "dobj":
        dObj = token
        tVerb = token.head
        print(tVerb, dObj)
        
        
# Extract he helper verb

intentVerb = None
verbList = ["want", "like", "need", "order"]

if tVerb.text in verbList:
    intentVerb = tVerb
    print(tVerb)
    
else:
    if tVerb.head.dep_ == "ROOT":
        helperVerb = tVerb.head
        intentVerb = helperVerb
        print(intentVerb)


# Extract the object of intent

intentObj = None
objList = ["flight", "meal", "booking"]

if dObj.text in objList:
    intentObj = dObj
    
else: 
    for child in dObj.children:
        
        if child.dep_ == "prep":
            intentObj = list(child.children)[0]
            break
        
        elif child.dep_ == "compound":
            intentObj = child
            break
            
    print(intentVerb.text+intentObj.text.capitalize())
```
